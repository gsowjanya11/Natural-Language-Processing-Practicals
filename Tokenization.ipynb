```python
from nltk.tokenize import sent_tokenize,word_tokenize,wordpunct_tokenize,TreebankWordTokenizer
```

```python
corpus="""This is sowjanya. I am learning Natural Language Processing. I am excited in this learning journey!!!"""
```

```python
sent=sent_tokenize(corpus,language="english")
print(sent)
```

```python
for l in sent:
    print(l)
```

```python
word_tokenize(corpus)
```

```python
sent
```

```python
for l in sent:
    print(word_tokenize(l))
```

```python
for l in sent:
    print(wordpunct_tokenize(l))
```

```python
wordpunct_tokenize(corpus)
```

```python
tokenizer=TreebankWordTokenizer()
tokenizer.tokenize(corpus)
```
