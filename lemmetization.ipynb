```python
from nltk.stem import WordNetLemmatizer
lemme=WordNetLemmatizer()
```

```python
'''
POS- Noun-n
verb-v
adjective-a
adverb-r
'''
print(lemme.lemmatize("eating",pos='n'))  ## Parts of Speech (POS - noun)
print(lemme.lemmatize('eating',pos='v'))  ## Parts of Speech (POS - verb)
print(lemme.lemmatize('eating',pos='a'))  ## Parts of Speech (POS - adjective)
print(lemme.lemmatize('eating',pos='r'))  ## Parts of Speech (POS - adverb)

print(lemme.lemmatize("goes",pos='n'))  ## Parts of Speech (POS - noun)
print(lemme.lemmatize('goes',pos='v'))  ## Parts of Speech (POS - verb)
print(lemme.lemmatize('goes',pos='a'))  ## Parts of Speech (POS - adjective)
print(lemme.lemmatize('goes',pos='r'))  ## Parts of Speech (POS - adverb)

print(lemme.lemmatize("beautiful",pos='n'))  ## Parts of Speech (POS - noun)
print(lemme.lemmatize('beautiful',pos='v'))  ## Parts of Speech (POS - verb)
print(lemme.lemmatize('beautiful',pos='a'))  ## Parts of Speech (POS - adjective)
print(lemme.lemmatize('beauty',pos='r'))  ## Parts of Speech (POS - adverb)

print(lemme.lemmatize("pouring",pos='n'))  ## Parts of Speech (POS - noun)
```

```python
data="""I am eating and walking . I would like to take some rest and have fun watching movies"""
data.lower()
from nltk.tokenize import word_tokenize
words=word_tokenize(data)
for l in words:
    print(f"The word {l}  ---> ",lemme.lemmatize(l,pos='n'))
```
