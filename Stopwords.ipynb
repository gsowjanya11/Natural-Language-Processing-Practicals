```python
import nltk
nltk.download('stopwords')
```

```python
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
from nltk.stem import PorterStemmer
stemmer=PorterStemmer()
```

```python
d="""I am able to do whatever i can till i breathe but my family encourages me to limit my abilities and start reading"""

w=word_tokenize(d)
```

```python
sentences=nltk.sent_tokenize(paragraph)
for l in sentences:
    print(l)
```

```python
w
' '.join(w)
```

```python
## Converting sentences into words and comparing words against stop words and stemming the final result
for i in range(len(sentences)):
    w=word_tokenize(sentences[i])
    w=[stemmer.stem(word) for word in w if word not in set(stopwords.words('english'))]
    sentences[i]=' '.join(w)
```
